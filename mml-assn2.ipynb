```python
import pandas as pd

# Define window size
window_size = 3  # Adjust this according to your preference

# Apply moving average to the 'Close' column
df['Close_smooth'] = df['Close'].rolling(window=window_size).mean()

# Replace NaN values with the mean of the 'Close_smooth' column
mean_close_smooth = df['Close_smooth'].mean()
df['Close_smooth'].fillna(mean_close_smooth, inplace=True)

# Print the dataframe with NaN values replaced
print(df)
```
